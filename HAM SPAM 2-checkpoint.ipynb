```python
#1. lower case
#2. tokenization
# remove special characters
#removing  stopwords and pucntuation
#stemming
```

```python
from nltk.corpus  import stopwords
stopword = stopwords.words('english')
```

```python
from  string import punctuation
p = list(punctuation)
```

```python
from nltk.stem.porter  import PorterStemmer
ps = PorterStemmer()
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    y = []
    for i in text:
        if  i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopword and i not in p:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
        
    return ' '.join(y)
```

```python
df1['message'] = df1['message'].apply(transform_text)
```
